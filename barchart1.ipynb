```python
Label=['A','B','C','D']
Values=[10,20,30,40]
bars=plt.bar(Label,Values)
plt.xlabel('Furniture')
bars[0].set_hatch('/')
plt.ylabel('Sales')
plt.title('Department Stores',color='g',fontsize='20')
plt.show()
```

```python
plt.barh(Label,Values)
plt.show()
```

```python
df=pd.DataFrame({'Male':[20,40,50,60],
                'Female':[23,12,90,34]},
                 index=['India','USA','Canada','UK'])
df.plot(kind="barh",title="Gender wise report",color=['green','blue'])
plt.xlabel('counts')
plt.ylabel('country')
plt.show()
```

```python
df=pd.DataFrame({'Male':[20,40,50,60],
                'Female':[23,12,90,34],
                'country':['India','USA','Canada','UK']})
#print(df)
plt.bar(df.country,df.Male,color='g')
plt.bar(df.country,df.Female,color='r',alpha=0.2)

plt.show()
```
