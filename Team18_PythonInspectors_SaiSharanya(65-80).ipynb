# Q66 Find the way to reverse string for given input (eg: input : this is python hackathon output:hackathon python is this)

# Q67 From'Python Hackthon Dec 2023' give 'Python Hack' using RegEX

```python
#import RegEX
import re

#searching the string for the word
input = "Python Hackthon Dec 2023"
match = re.search(r'Python Hack', input)

# Based on the match printing the output
if match:
    result = match.group()
    print(result)
else:
    print("No match found.")
```

# Q73 Find the max EDA of the higest scoring student (whose average of midterm1,midterm2 and Final is maximum)

```python
import pandas as pd
# Calculating avg of Midterm1, Midterm2 and final
grades['Average'] = grades[['Midterm 1', 'Midterm 2', 'final']].mean(axis=1)
merged_df = pd.merge(grades, eda, on='Student_id')

#Calculating Highest scoring student
max_average = merged_df['Average'].max()
max_grades_data = merged_df[merged_df['Average'] == max_average]
print(max_grades_data)

#Max EDA of the highest scoring student 
max_eda = max_grades_data['EDA'].max()
print("Maximum EDA of the highest scoring student", max_eda)
```

# Q74 Create a Pie chart for  Grades and explode the wedge with Highest Grade in Final and Label with the Student ID

```python
import pandas as pd
import matplotlib.pyplot as plt

# Find the student with the highest final grade
highest_grade_student = grades.loc[grades['final'].idxmax(), 'Student_id']

# Create a pie chart
labels = grades['Student_id']
sizes = grades['final']

# Explode the wedge with the highest grade
explode = [0.1 if student == highest_grade_student else 0 for student in grades['Student_id']]

# Plotting the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True, startangle=90)
plt.title('Final Grades Distribution')
plt.show()
```

# Q75 Find min temparature of the student whose average score (midterm1,midterm2, final) is minimum

```python
#Average of Midterm1, Midterm2 and final 
grades['Average'] = grades[['Midterm 1', 'Midterm 2', 'final']].mean(axis=1)
merged_df = pd.merge(grades, temp, on='Student_id')

#Minimum value of datasetgrades
min_average = merged_df['Average'].min()
min_grades_data = merged_df[merged_df['Average'] == min_average]
print(min_grades_data)


min_temp = min_grades_data['TEMP'].min()
print("Minimum temp of the min scoring student", min_temp)

```

# Q80 Create a Regression Plot on Max EDA and mean HR 

```python
merged_df = pd.merge(eda, hr, on='Student_id')
max_eda_value=merged_df.groupby('Student_id')['EDA'].max().reset_index()
mean_hr_value=merged_df.groupby('Student_id')['HR'].mean().reset_index()

print(max_eda_value)
plt.figure(figsize=(8, 6))
sns.regplot(x=max_eda_value['EDA'], y=mean_hr_value['HR'], data=merged_df, scatter_kws={'s':100})
plt.title('Regression Plot: Max EDA vs. Mean HR')
plt.xlabel('Max EDA')
plt.ylabel('Mean HR')
plt.grid(True)
plt.show()
```
